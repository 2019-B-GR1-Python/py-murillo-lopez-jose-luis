```python
import numpy as np
from scipy import ndimage
from scipy import misc
```

```python
mapache = misc.face()
mapache
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(mapache)
plt.show()
```

```python
plt.imshow(mapache[::-1])
plt.show()
```

```python
mapache1, mapache2 = np.split(mapache, 2)
```

```python
plt.imshow(mapache1)
plt.show()
```

```python
plt.imshow(mapache2)
plt.show()
```

```python
mapache.shape
```

```python
#1. Hacer al mapache azul
blue = np.array((0,0,255))
blue
```

```python
blue[None,None,:]
```

```python
mapache_azul = mapache + blue[None,None,:]
mapache_azul = mapache_azul / 510
mapache_azul
```

```python
plt.imshow(mapache_azul)
plt.show()
```

```python
#Alterar el espacio-tiempo en el mapache (rompecabezas)
arreglo_random = np.arange(8)
np.random.shuffle(arreglo_random)
arreglo_random
```

```python
mapache_rp_1 = np.hsplit(mapache1,4)
mapache_rp_2 = np.hsplit(mapache2,4)
mapache_rp = np.array(mapache_rp_1 + mapache_rp_2)
mapache_rp
```

```python
np.random.shuffle(mapache_rp)
```

```python
mapache_alterado_1 = mapache_rp[0]
for i in range(1,4):
    mapache_alterado_1 = np.concatenate((mapache_alterado_1, mapache_rp[i]),1)
mapache_alterado_2 = mapache_rp[4]
for i in range(5,8):
    mapache_alterado_2 = np.concatenate((mapache_alterado_2, mapache_rp[i]),1)
mapache_alterado_2.shape
mapache_alterado = np.concatenate((mapache_alterado_1, mapache_alterado_2), 0)
```

```python
plt.imshow(mapache_alterado)
plt.show()
```

```python
#Oscurecer la imagen
mapache_oscuro = mapache / 765

plt.imshow(mapache_oscuro)
plt.show()
```

```python
#Hacer monocromatico al mapache
mono = np.array((0.2125, 0.7154, 0.0721))

mapache_mono = mapache / 255

mapache_mono = mapache_mono * mono

mapache_mono = mapache_mono.sum(axis=2)

plt.imshow(mapache_mono)
plt.show()
```

```python
#Aplicar un filtro gaussiano
mapache_gaussiano = ndimage.gaussian_filter(mapache,5)

plt.imshow(mapache_gaussiano)
plt.show()
```
